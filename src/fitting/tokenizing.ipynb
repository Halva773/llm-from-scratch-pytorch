```python
import pandas as pd
```

```python
import sys  
sys.path.insert(1, '..')
from bpe import BPE
```

```python
filepath = "../../dataset/poems.csv"
data = pd.read_csv(filepath)
data = data.dropna(subset=['text'])
print(data.shape)
data.head(2)
```

```python
data.loc[:1000, 'text'].str.len().sum()
```

```python
%%time

bpe = BPE(vocab_size=500)
whole_text = "\n".join(data.loc[:1000, 'text'])
print("Text formatted. Going to fit tokenizer")
bpe.fit(whole_text)
bpe.save('../savepoints/bpe.dill')
```
